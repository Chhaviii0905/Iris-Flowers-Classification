# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size = 0.3)
```

```python
train.shape
```

```python
test.shape
```

```python
X_train = train[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
                 'PetalWidthCm']]
y_train = train.Species

X_test = test[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm',
                 'PetalWidthCm']]
y_test = test.Species
```

```python
X_train.head()
```

```python
y_test.head()
```

# Modelling and Accuracy

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1 = LogisticRegression()
model1.fit(X_train, y_train)
pred1 = model1.predict(X_test)
print('Accuracy:', accuracy_score(pred1,y_test))
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, pred1)
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, pred1, labels=model1.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels=model1.classes_)
disp.plot()
print("TN: ", cm[0][0])
print("FN: ", cm[1][0])
print("TP: ", cm[1][1])
print("FP: ", cm[0][1])
```
